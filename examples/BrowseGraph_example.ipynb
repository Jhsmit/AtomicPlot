```python
%matplotlib notebook
import matplotlib.pyplot as plt
from atomicplot.ipython import BrowseGraph
import numpy as np
```

```python
data1 = np.random.rand(200, 300)
data2 = np.random.normal(size=(400, 200))
```

```python
def plot_func(data, figure):
    ax1, ax2 = figure.axes
    d1, d2 = data
    ax1.plot(d1)
    ax2.plot(d2)
    plt.tight_layout()
```

```python
fig, ax = plt.subplots(2, 1)
BrowseGraph(list(zip(data1, data2)), plot_func, fig)
```
